```python
import matplotlib.pyplot as plt
import os
```

```python
class Set:
    def __init__(self, heights, accuracies, index):
        self.heights = heights
        self.accuracies = accuracies
        self.index = index

        self.best_height = self.__find_best_height()

    def __find_best_height(self):
        max_acc = 0
        best_height = 0

        for i in range(len(self.heights)):
            if self.accuracies[i] > max_acc:
                best_height = self.heights[i]
                max_acc = self.accuracies[i]
        
        return best_height
```

```python
def read_file(inf, set_index):
    lines_count = int(inf.readline())

    heights = []
    accuracies = []
    for i in range(lines_count):
        line = inf.readline().split()

        heights += [int(line[0])]
        accuracies += [float(line[1])]

    return Set(heights, accuracies, set_index)
```

```python
def read_data(folder_path):
    files = os.listdir(folder_path)

    sets = []
    for i in range(len(files)):
        with open(folder_path + '\\' + files[i]) as inf:
            sets += [read_file(inf, i)]

    return sets
```

```python
sets = read_data("res")
```

```python
min_height_set = 0
max_height_set = 0

for i in range(len(sets)):
    if sets[i].best_height < sets[min_height_set].best_height:
        min_height_set = i
    
    if sets[i].best_height > sets[max_height_set].best_height:
        max_height_set = i
```

```python
print("Set with min best height is ", min_height_set + 1)
print("Set with max best height is ", max_height_set + 1)

_, chart = plt.subplots()

chart.plot(sets[min_height_set].heights, sets[min_height_set].accuracies, label = "Min best height")
chart.plot(sets[max_height_set].heights, sets[max_height_set].accuracies, label = "Max best height")

chart.legend()
chart.set_xlabel("Height")
chart.set_ylabel("Accuracy")

plt.show()
```
